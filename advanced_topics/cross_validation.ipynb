```python
import pandas as pd

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

Create a feature named gt_avg, which should be either 1 or 0. The value should indicate whether or not a given price is greater than the average price for that car's combination of year, make, and model.


```python
df['avg_price'] = df.groupby(['Year', 'Make', 'Model']).Price.transform('mean')

df['gt_avg'] = df.Price > df.avg_price

df['gt_avg'] = LabelEncoder().fit(df.gt_avg).transform(df.gt_avg)

df[['Price', 'avg_price', 'gt_avg']].sample(10)

```

Drop the Id, City, and Vin columns.


```python
df.drop(columns=['Id', 'City', 'Vin'], inplace=True)
```

```python
df.head()
```

Encode the categorical features as necessary. You might wish to use a sklearn.preprocessing.LabelEncoder.


```python
df['Year'] = LabelEncoder().fit(df.Year).transform(df.Year)
df['State'] = LabelEncoder().fit(df.State).transform(df.State)
df['Make'] = LabelEncoder().fit(df.Make).transform(df.Make)
df['Model'] = LabelEncoder().fit(df.Model).transform(df.Model)
```

```python
df[['Price', 'avg_price', 'gt_avg']].head(10)
```

Split the data into training and test sets.

```python
X = df.drop(columns=['Price', 'avg_price', 'gt_avg'])
y = df[['gt_avg']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 123)
```
